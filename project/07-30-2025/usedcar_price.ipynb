```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import RobustScaler
from sklearn.decomposition import PCA
```

```python
car_df = pd.read_csv('cars.csv')
brand_df = pd.read_csv('brand.csv')
```

```python
car_df.head()
```

```python
brand_df.head()
```

```python
# car_df 와 brand_df 합치기
# title 값의 첫 번째 단어만 추출하여 brand 칼럼에 저장
# object series 이므로, `.str` 같은 accessor를 통해 모든 열에 적용

car_df['brand'] = car_df['title'].str.split().str[0]      
```

```python
car_df.head()
```

```python
# brand_df의 title 칼럼에 저장된 값들을 대문자로 변환

brand_df['title'] = brand_df['title'].str.upper()
```

```python
brand_df.head()
```

```python
# car_df의 brand 칼럼과 brand_df의 title 칼럼을 기준으로 데이터를 병합
# car_df와 brand_df 의 칼럼명이 다르므로, left_on 과 right_on 활용

car_df = car_df.merge(brand_df, left_on = 'brand', right_on = 'title', how = 'left')
```

```python
car_df.head()
```

```python
# car_df 의 중복된 title_y drop

car_df.drop('title_y', axis = 1, inplace = True)
```

```python
# car_df 의 title_x > title 변경

car_df.rename({'title_x': 'title'}, axis = 1, inplace = True)
```

```python
car_df.head()
```

```python
# car_df의 복사본 bonus_df를 생성합니다.

bonus_df = car_df.copy()
```

```python
car_df.info()
```

```python
# Engine 칼럼의 데이터에서 L 문자를 제거하여 숫자 부분만 저장

car_df['Engine'] = car_df['Engine'].str.rstrip('L')
```

```python
# Emission Class 칼럼 데이터에서 숫자 부분만 저장

car_df['Emission Class'] = car_df['Emission Class'].str.split().str[1]
```

```python
# to_numeric 메서드를 사용하여 Engine, Emission Class 컬럼의 자료형을 숫자 형태로 변환

car_df['Engine'] = pd.to_numeric(car_df['Engine'])
car_df['Emission Class'] = pd.to_numeric(car_df['Emission Class'])
```

```python
# 특정 칼럼들이 몇 종류의 데이터로 이루어져 있는지 계산
# 범주형 칼럼을 식별

# title 고유값 448개로 drop 하거나 brand 대체
# one-hot encoding 혹은 label encoding 활용

car_df[['title','Fuel type','Body type','Gearbox','Emission Class','Service history','brand','country']].nunique()
```

```python
# title drop

car_df.drop('title', axis = 1, inplace = True)
```

```python
# brand 데이터 개수 계산

car_df['brand'].value_counts()
```

```python
# 브랜드별 평균 가격 계산.

car_df.groupby('brand')['Price'].mean()
```

```python
# 브랜드별 데이터 개수와 평균 가격 출력 - pd.concat() 활용

brand_count = car_df['brand'].value_counts()
brand_price = car_df.groupby('brand')['Price'].mean()

pd.concat([brand_count, brand_price], axis = 1)
```

```python
# 브랜드별 데이터 개수와 평균 가격 출력 - to_frame() & join() 활용

brand_count = car_df['brand'].value_counts()
brand_price = car_df.groupby('brand')['Price'].mean()

# dtype = series 이므로 DataFrame 으로 변경

brand_temp = brand_count.to_frame().join(brand_price.to_frame())

brand_temp
```

```python
# 범주형 칼럼에 one-hot encoding

car_df = pd.get_dummies(car_df, drop_first = True)
```

```python
# Robust Scaler

rs = RobustScaler()
```

```python
# robust scaling을 적용한 데이터프레임을 car_df 변수에 다시 저장

car_df = pd.DataFrame(rs.fit_transform(car_df), columns = car_df.columns)
```

```python
# 주성분이 5개인 PCA 모델 정의

pca = PCA(5)
```

```python
# car_df 에 pca 적용

pd.DataFrame(pca.fit_transform(car_df))
```

```python
# 원본 데이터 대비 PCA 데이터의 분산(explained variance) 비율 계산

pca.explained_variance_ratio_
```

```python
# 주성분의 개수가 2~10개일 때 explained variance ratio를 각각 계산

for i in range(2, 11):
    pca = PCA(i)
    pca.fit(car_df)
    print(i, round(pca.explained_variance_ratio_.sum(), 2))
```

```python
# 누적 설명력 7~80% 이상
# 차원 축소 및 정보 손실 최소화 주성분 개수 선택

pca = PCA(7)
```

```python
# PC1~PC7 컬럼을 출력

pd.DataFrame(pca.fit_transform(car_df), columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7'])
```
