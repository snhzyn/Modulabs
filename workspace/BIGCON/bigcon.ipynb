### RANK 전처리

```python
df_2.columns
```

```python
import re

# 앞자리 추출
def find_rank(x):
    if pd.isna(x):
        return None
    match = re.search(r'\d+', str(x))
    
    return int(match.group()) if match else None

df_2_rank_cols = [
    'MCT_OPE_MS_CN', 
    'RC_M1_SAA', 
    'RC_M1_TO_UE_CT',
    'RC_M1_UE_CUS_CN', 
    'RC_M1_AV_NP_AT', 
    'APV_CE_RAT'
]

for col in df_2_rank_cols:
    df_2[col] = df_2[col].apply(find_rank)

df_2.head()
```
